# Find the initial values for Ex 29 using bisection.

-  Find  a  necessary  condition  that  a  curve  intersects  with  a  rectangle.   Use  the function value in the center, and bounds for the gradient of $f$ on the rectangle.
-  start with a user-provided rectangle
-  recursively sub-divide the rectangle into 4 smaller ones as long as both curves can intersect with the rectangle. If the rectangle is smaller than a user-provided value use Newtonâ€™s method of Ex 29, with the center of the rectangle as initial guess.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from scipy.spatial import distance

class quadratic_curve:
    def __init__(self, A, b, c):
        self.A = A
        self.b = b
        self.c = c
        
    def evaluate(self, xy):
        return xy.T @ self.A @ xy + self.b.T @ xy + self.c
    
    def evaluate_gradient(self, xy):
        x, y = xy
        
        dx = 2*self.A[0,0]*x + 2*self.A[0,1]*y + self.b[0]
        dy = 2*self.A[1,1]*y + 2*self.A[0,1]*x + self.b[1]
        return np.array([dx, dy])
    
    def plot(self, x_min, x_max, y_min, y_max):
        x = np.linspace(x_min, x_max, 100)
        y = np.linspace(y_min, y_max, 100)
        X, Y = np.meshgrid(x, y)
        
        Z = self.A[0,0]*X**2 + 2*self.A[0,1]*X*Y + self.A[1,1]*Y**2 + self.b[0]*X + self.b[1]*Y + self.c
        
        plt.contour(X, Y, Z, [0])
        

def curve_in_box(c, p1, p2):
    center_point = 0.5 * (p1 + p2)
    box_dimensions = np.abs(p2 - p1)
    
    #print(center_point)
    #print(box_dimensions)
    
    center_value = c.evaluate(center_point)
    center_gradient = c.evaluate_gradient(center_point)
    
    #print(center_value)
    #print(center_gradient)
    
    d = np.linalg.norm(box_dimensions, ord=2)
    f_norm = np.linalg.norm(center_gradient, ord=2)
    H_norm = 2*np.linalg.norm(c.A, ord=2)
    
    #print(d)
    #print(f_norm)
    #print(H_norm)
    
    return np.abs(center_value) < (d * f_norm + 2*H_norm*d)/10
    
        
def bisect(c1, c2, p1, p2, iter_idx, epsilon=1e-2, plotting=False):
    #if iter_idx > 5:
        #return
        #pass
    
    center_point = 0.5 * (p1 + p2)
    box_dimensions = np.abs(p2 - p1)[np.newaxis].T
       
    return_list = []    
    
    if curve_in_box(c1, p1, p2) and curve_in_box(c2, p1, p2):
        if np.max(box_dimensions) < epsilon:
            return [center_point]
        
        p1_list = p1[np.newaxis].T + box_dimensions/2 * [[0, 0, 1, 1], [0, 1, 0, 1]]
        p2_list = p2[np.newaxis].T - box_dimensions/2 * [[1, 1, 0, 0], [1, 0, 1, 0]]
        for sub_idx in range(4):
            if plotting:
                rect = patches.Rectangle(p1_list[:, sub_idx],box_dimensions[0]/2,
                                         box_dimensions[1]/2,linewidth=1,edgecolor='r',facecolor='none')
                
                ax = plt.gca()
                ax.add_patch(rect)
                
            return_val = bisect(c1, c2, p1_list[:, sub_idx], p2_list[:, sub_idx], iter_idx+1, 
                                epsilon=epsilon, plotting=True)
            
            if return_val is not None and len(return_val) > 0:
                return_list = return_list + return_val
    
    if len(return_list) > 0:
        return return_list
    

def intersection_curve(c1, c2, starting_point, max_iter=30, epsilon=1e-10):
    current_point = starting_point
    point_list = [starting_point]

    for iter_idx in range(max_iter):
        value_c1 = c1.evaluate(current_point)
        value_c2 = c2.evaluate(current_point)
        current_f = np.vstack((value_c1, value_c2))

        grad_c1 = c1.evaluate_gradient(current_point)
        grad_c2 = c2.evaluate_gradient(current_point)

        jacobi = np.vstack((grad_c1.T, grad_c2.T))

        updated_point = current_point - np.linalg.inv(jacobi) @ current_f
        point_list.append(updated_point)

        if np.linalg.norm(updated_point - current_point) < epsilon:
            return updated_point, point_list

        current_point = updated_point

    return None, point_list
```

```python
A = np.array([[2.7, 0.8], [0.8, 0.8]])
b = np.array([[-1.5], [1.6]])
c = -0.6

c1 = quadratic_curve(A, b, c)

A = np.array([[1.7, 0.9], [0.9, -0.5]])
b = np.array([[0.2], [-1.9]])
c = -1.1

c2 = quadratic_curve(A, b, c)


plt.figure(figsize=(13, 9))
c1.plot(-2, 3, -4, 3)
c2.plot(-2, 3, -4, 3)

# Create a Rectangle patch
rect = patches.Rectangle((-2,0),1,1,linewidth=1,edgecolor='r',facecolor='none')

# Add the patch to the Axes
ax = plt.gca()
#ax.add_patch(rect)
```

```python
plt.figure(figsize=(13, 9))
c1.plot(-2, 3, -6, 3)
c2.plot(-2, 3, -6, 3)
possible_starting_points = bisect(c1, c2, np.array([-1.5, -5]), np.array([2.5, 2.5]), 0, plotting=False)

```

```python
coords = np.array(possible_starting_points)
dists = distance.cdist(coords, coords, 'euclidean')

mask = np.triu(dists < 1e-2 * 1.5)
mask = sum(mask)==1
starting_pts = coords[mask, :]
```

```python
intersection_pt_lists = []
for starting_pt in starting_pts:
    intersection_pt, __ = intersection_curve(c1, c2, starting_pt[np.newaxis].T, max_iter=30, epsilon=1e-10)
    
    intersection_pt_lists.append(intersection_pt)
    
intersection_pt_lists = np.array(intersection_pt_lists).squeeze()
print(intersection_pt_lists)
```

```python
plt.figure(figsize=(13, 9))
c1.plot(-2, 3, -6, 3)
c2.plot(-2, 3, -6, 3)
plt.plot(intersection_pt_lists[:, 0], intersection_pt_lists[:, 1], 'ro')
```
