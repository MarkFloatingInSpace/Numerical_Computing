# Implement a function for numerical integration by a composite integration rule

Input is the function to be integrated, the interval $[a,b]$, the amount of subintervals $m$, and aquadrature rule $[(c0,ω0),...,(cn,ωn)]$ with reference points $c_i ∈[0,1]$ and weights $ω_i$.

Study convergence for the functions $f(x) =\int_0^1 e^x dx$ and $f(x) =\sqrt x$, and integrationrules

-  (a)  trapezoidal:  $[(0,0.5),(1,0.5)]$
-  (b)  Simpson:  $[(0,1/6),(0.5,2/3),(1,1/6)]$
-  (c)  mid-point:  $[(0.5,1)]$
-  (d)  2-point Gaussian rule:  $[(0.5−0.5/√3,0.5),(0.5 + 0.5/√3,0.5)]$

### Idea:
Approximation of the integral of a function $f(x)$ with the integral of a polynomial interpolation:

$$ \int_a^b f(x) dx \approx \int_a^b I_f(x) dx =: Q(f)$$

with the polynomial written w.r.t the Lagrange basis:

$$ I_f(x) = \sum_{j=0}^n f(x_j) l_j(x)$$

$$ \int_a^b I_f(x) dx = \int_a^b \sum_{j=0}^n f(x_j) l_j(x) dx = \sum_{j=0}^n \int_a^b l_j(x) dx f(x_j)$$

with the intergration weight
$$ \omega_j = \int_a^b l_j(x) dx $$
$$ \sum_{j=0}^n \omega_j f(x_j) $$

By always transforming the interval $[a, b]$ to the unit interval $[0, 1]$ the normalized reference wieghts and points can be calculated once and for all.

$$ \int_a^b f(x) dx = \int_0^1 f(a+sh) h  ds$$
with $h$ being the interval length.

Therefore:
$$ \int_a^b f(x) dx \approx h \sum_{j=0}^n \omega_j f(a+c_j h)$$
with $c_j$ and $\omega_j$ being normalized reference weights and points.

For better apprxomation performance the interval $[a, b]$ can be split into $m$ sub intervals.

```python
import numpy as np
import matplotlib.pyplot as plt


def composite_integration(f, a, b, m, quad_rule):
    c_k = np.array([quad_rule[i][0] for i in range(len(quad_rule))])
    w_k = np.array([quad_rule[i][1] for i in range(len(quad_rule))])
    
    h = (b-a)/m
    
    total_integral = 0
    for sub_intervall_idx in range(m):
        x_j = a + sub_intervall_idx * h
        
        f_values = f(x_j + c_k*h)
        integral_subinterval = h * np.sum(f_values * w_k)
        
        total_integral += integral_subinterval
        
    return total_integral


def fun1(x):
    return np.exp(x)


def fun2(x):
    return np.sqrt(x)
    
    
fun1_true_val = np.e - 1
fun2_true_val = 2/3
    
trapezoidal_rule = [(0,0.5),(1,0.5)]
simpson_rule = [(0,1/6),(0.5,2/3),(1,1/6)]
midpoint = [(0.5,1)]
two_pt_gaussian_rule = [(0.5-0.5/np.sqrt(3),0.5),(0.5 + 0.5/np.sqrt(3),0.5)]

rule_list = [trapezoidal_rule, simpson_rule, midpoint, two_pt_gaussian_rule]
rule_name_list = ["trapezoidal rule", "simpson rule", "mid-point", "2-point Gaussian rule"]

max_m = 1000

error_matrix = np.empty((max_m, len(rule_list)))
error_matrix[:] = np.nan
for rule_idx in range(len(rule_list)):
    for m in range(1, max_m+1):
        error_matrix[m-1, rule_idx] = np.abs(composite_integration(fun1, 0, 1, m, rule_list[rule_idx]) - fun1_true_val)
        
plt.figure(figsize=(12, 12))
plt.plot(error_matrix)
plt.legend(rule_name_list)
plt.yscale("log")
plt.grid()

error_matrix = np.empty((max_m, len(rule_list)))
error_matrix[:] = np.nan
for rule_idx in range(len(rule_list)):
    for m in range(1, max_m+1):
        error_matrix[m-1, rule_idx] = np.abs(composite_integration(fun2, 0, 1, m, rule_list[rule_idx]) - fun2_true_val)
        
plt.figure(figsize=(12, 12))
plt.plot(error_matrix)
plt.legend(rule_name_list)
plt.yscale("log")
plt.grid()
```
